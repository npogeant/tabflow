```python
import sys
from pathlib import Path

import pandas as pd
import openpyxl

sys.path.append(str(Path.cwd() / "src"))

from pageflow import PageRange
```

```python
df = pd.DataFrame({
    "Store": ["Store 1"],
    "Unique Staff Code": ["12345"],
    "Staff Name": ["John Doe"],
    "Transactions Bags": [45],
    "Transactions Bags and Cross Sell": [32],
    "% Cross Sell": [70.96]
})

df
```

```python
# open workbook and get named range data
wb = openpyxl.load_workbook("test.xlsx")
```

```python
PageRange(wb, 'PERIOD').update_single_cell(f"From today to a week from now")
PageRange(wb, 'DATA').set_data(df, has_header=True)
```

```python
wb.save("test_updated.xlsx")
```
